```python
import supersuit as ss
from stable_baselines3 import PPO
from stable_baselines3.ppo import CnnPolicy
from stable_baselines3.ppo import MlpPolicy
from pettingzoo.sisl import multiwalker_v9


env = multiwalker_v9.env()
env = ss.frame_stack_v1(env, 3)

model = PPO.load("multiwalker_v9_policy")
records1 = []
for i in range(400):
    done = 0    
    env.reset()
    total = 0
    while not done:
        for agent in env.agent_iter():
            obs, reward, done, info = env.last()
            act = model.predict(obs, deterministic=True)[0] if not done else None
            env.step(act)
            total += reward
            env.render()
    records1.append(total)
    print('episode:',i,'total reward:',total)
```

```python
env = multiwalker_v9.env()
env = ss.frame_stack_v1(env, 3)

model = PPO.load("multiwalker_v9_policy")
records2 = []
for i in range(400):
    done = 0    
    env.reset()
    total = 0
    while not done:
        for agent in env.agent_iter():
            obs, reward, done, info = env.last()
            act = env.action_space(agent).sample()
            env.step(act)
            total += reward
    records2.append(total)
    print('episode:',i,'total reward:',total)
```

```python
# 导入 pandas 模块并将其简化为 pd
import pandas as pd
# 导入pyplot模块
from matplotlib import pyplot as plt

data_list = pd.DataFrame({ 'policy-1':records1,'policy-2':records2}) #将列表转化为DataFram数据型,records1/2即为两次模型的数据列表

# 生成画布，并设置画布的大小为(6, 6)
plt.figure(figsize=(6, 6))

# 设置 x/y坐标值
x = data_list.index
y = data_list.values

# 绘制折线图
plt.plot(x, y,)

# 设置图表标题名为'Line chart of rewards of policys'，字体大小为 20
plt.title('Line chart of rewards of policys',fontsize = 20)

# 设置坐标轴的刻度字体大小为 12
plt.xticks(fontsize =12)
plt.yticks(fontsize =12)

# 设置 x、y 轴标题分别为'expoide'、'total reward'，字体大小均为 15
plt.xlabel('expoide',fontsize = 15)
plt.ylabel('total reward',fontsize = 15)

# 设置图例为'policy-1'、'policy-2'
plt.legend(['policy-1','policy-2'])

plt.xlim((0,450))  #自定义x轴范围0，450
plt.show()
```
